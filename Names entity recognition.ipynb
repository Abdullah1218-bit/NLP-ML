```python
corpus='Apple Inc. announced on December 1, 2025, that it will open a new office in Berlin, Germany. The CEO, Tim Cook, stated that the expansion will create over 500 jobs in the city. Meanwhile, Microsoft Corporation is planning to launch its latest software update in New York next month. In sports news, Lionel Messi scored two goals for Paris Saint-Germain in the Champions League match against Real Madrid at the Santiago Bernabéu Stadium'
import nltk
words=nltk.word_tokenize(corpus)
words
```

```python
from nltk import pos_tag
tag_elements=pos_tag(words)
tag_elements
```

```python
nltk.ne_chunk(tag_elements)
```

```python
nltk.ne_chunk(tag_elements).draw()
```
