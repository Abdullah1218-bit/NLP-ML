```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
wnl.lemmatize('going',pos='v')
```

```python
words=['program','programming','programmar','programs','write','wrote','written','writes',
       'eat','ate','eaten','eating','eats','history']
for word in words:
    print(word+'---->'+wnl.lemmatize(word,pos='v'))
```

```python
words_2=['university','policy','analysis','organization','hapiness','fairly','supportingly']
for word in words_2:
    print(word+'---->'+wnl.lemmatize(word,pos='v'))
```
