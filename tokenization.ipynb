```python
corpus='''Natural language processing is a field of artificial intelligence. It focuses on the interaction between computers and humans. Machine learning helps systems learn from data. I like it.
Hello's and Hey`s

'''
corpus
```

# ***Sent Tokenize***

```python
from nltk.tokenize import sent_tokenize
document=sent_tokenize(corpus)
document
```

# ***Word Tokenize***

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

# ***TreebankWordTokenizer***

```python
from nltk.tokenize import TreebankWordTokenizer
token=TreebankWordTokenizer()
token.tokenize(corpus)
```
