```python
words=['program','programming','programmar','programs','write','wrote','written','writes',
       'eat','ate','eaten','eating','eats','history']
words_2=['university','policy','analysis','organization','hapiness','fairly','supportingly']
```

# ***Porter Stemmmer***

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
for word in words:
    print(word + '----->'+stemming.stem(word))
```

```python
stemming=PorterStemmer()
for word in words_2:
    print(word + '----->'+stemming.stem(word))
```

# ***Regular Expression Stemmmer***


```python
from nltk.stem import RegexpStemmer
st = RegexpStemmer('ing$|s$|e$|able$', min=4)
st.stem('caring'),st.stem('care'),st.stem('ablecaring')
```

# ***SnowballStemmer***


```python
from nltk.stem import SnowballStemmer
ss=SnowballStemmer('english')

for word in words_2:
    print(word + '----->'+ss.stem(word))
```
